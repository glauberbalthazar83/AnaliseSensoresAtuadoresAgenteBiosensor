```python
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
import seaborn as sns
py.init_notebook_mode(connected=True)
%matplotlib inline
```

```python
dados  = pd.read_csv("/home/balthazar/ARTIGO1/dados_temp_umid.csv",sep=";")
dados_temp = dados[0:36]
dados_umid = dados[36:]
```

```python
temp_ambiente = dados[0:12]
temp_aprox_objeto_quente = dados[12:24]
temp_aprox_objeto_frio = dados[24:36]
umid_aprox_objeto_molhado = dados[36:48]
umid_ambiente = dados[48:]
```

```python
ax1 = temp_ambiente.plot(kind='scatter', x='Contador', y='BME_Temp', color='r', label='BME280') 
ax1.set_title(label="Temperatura: ambiente")
ax2 = temp_ambiente.plot(kind='scatter', x='Contador', y='DHT11_Temp', color='g', ax=ax1, label='DHT11')    
ax3 = temp_ambiente.plot(kind='scatter', x='Contador', y='DHT22_Temp', color='b', ax=ax1, label='DHT22')  
ax4 = temp_ambiente.plot(kind='scatter', x='Contador', y='P10_temp', color='y', ax=ax1, label='P10')  
ax5 = temp_ambiente.plot(kind='scatter', x='Contador', y='HTU21d_temp', color='c', ax=ax1, label='HTU21d')  
ax5.set_ylabel('Celsius')
ax5.set_xlabel('Registros')
print(ax1 == ax2 == ax3 == ax4 == ax5)
```
