```python
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
import seaborn as sns
py.init_notebook_mode(connected=True)
%matplotlib inline
```

```python
dados  = pd.read_csv("/home/balthazar/ARTIGO1/dadoslux.csv",sep=";")
```

```python
claridade_ambiente = dados[0:12]
claridade_forcada = dados[12:24]
penumbra = dados[24:36]
escuridao = dados[36:]
```

```python
ax1 = claridade_ambiente.plot(kind='scatter', x='Contador', y='TEMT6000', color='r', label='TEMT6000') 
ax1.set_title(label="Claridade (ambiente)")
ax2 = claridade_ambiente.plot(kind='scatter', x='Contador', y='BH1750', color='g', ax=ax1, label='BH1750')    
ax3 = claridade_ambiente.plot(kind='scatter', x='Contador', y='P13LDR', color='b', ax=ax1, label='P13LDR')  
ax3.set_ylabel('Lux')
ax3.set_xlabel('Registros')
print(ax1 == ax2 == ax3)
```

```python
ax1 = claridade_forcada.plot(kind='scatter', x='Contador', y='TEMT6000', color='r', label='TEMT6000') 
ax1.set_title(label="Claridade (for√ßada)")
ax2 = claridade_forcada.plot(kind='scatter', x='Contador', y='BH1750', color='g', ax=ax1, label='BH1750')    
ax3 = claridade_forcada.plot(kind='scatter', x='Contador', y='P13LDR', color='b', ax=ax1, label='P13LDR')  
ax3.set_ylabel('Lux')
ax3.set_xlabel('Registros')
print(ax1 == ax2 == ax3)
```

```python
ax1 = penumbra.plot(kind='scatter', x='Contador', y='TEMT6000', color='r', label='TEMT6000') 
ax1.set_title(label="Penumbra")
ax2 = penumbra.plot(kind='scatter', x='Contador', y='BH1750', color='g', ax=ax1, label='BH1750')    
ax3 = penumbra.plot(kind='scatter', x='Contador', y='P13LDR', color='b', ax=ax1, label='P13LDR')  
ax3.set_ylabel('Lux')
ax3.set_xlabel('Registros')
print(ax1 == ax2 == ax3)
```

```python
datasetClaridadeAmbiente = {
    'TEMT6000' : claridade_ambiente.TEMT6000,
    'BH1750': claridade_ambiente.BH1750,
    'P13LDR':claridade_ambiente.P13LDR
}
datasetClaridadeAmbiente = pd.DataFrame(datasetClaridadeAmbiente)

ax = sns.boxplot(data=datasetClaridadeAmbiente, orient='h', width=0.5)
ax.figure.set_size_inches(12,6)
ax.set_title('Claridade (ambiente)', fontsize=20)
ax.set_xlabel('Lux', fontsize=16)
ax.set_ylabel('Sensores', fontsize=16)
ax
```
